# **1. Análise dos clusters (interpretando as médias)**

Vamos traduzir a tabela:

| Cluster       | Características principais                                                       |
| ------------- | -------------------------------------------------------------------------------- |
| **Cluster 0** | Açúcar residual MUITO alto, sulfatos baixos, álcool e qualidade médios           |
| **Cluster 1** | Maior acidez volátil, maior teor de cloretos, menor açúcar residual, álcool alto |
| **Cluster 2** | Baixa densidade, álcool alto, qualidade mais alta                                |
| **Cluster 3** | Açúcar residual alto, sulfatos mais baixos, SO₂ intermediário                    |

Agora vamos detalhar:

---

# **Cluster 0 – Vinhos doces / mais suaves**

* **Maior residual_sugar (9.31)**
* **Maior total_sulfur_dioxide (198)**
* **Densidade mais alta** → típico de vinhos doces
* **Álcool mais baixo (9.7)**
* **Qualidade média (5.55)**

 **Interpretação:**
Este cluster representa vinhos **mais doces**, com **mais conservantes (SO₂)**, **menos alcoólicos** e qualidade **inferior**. Geralmente vinhos mais baratos e suaves.

---

# **Cluster 1 – Vinhos mais ácidos e alcoólicos**

* **Maior volatile_acidity (0.49)** → vinhos mais agressivos no aroma
* **Maior cloretos (0.081)**
* **Menor açúcar residual (2.47)**
* **Álcool alto (10.56)**
* **pH mais alto (3.29)**
* **Baixa densidade**

**Interpretação:**
Vinhos **secos**, mais alcoólicos, porém com **acidez volátil elevada**, o que prejudica o aroma e pode diminuir a qualidade.
Mesmo assim a qualidade média é **5.67**.

---

# **Cluster 2 – Vinhos mais equilibrados e de melhor qualidade**

* **Álcool mais alto (10.96)**
* **Baixa densidade (0.993)**
* **Sulfatos bons (0.50)**
* **Chlorides menores (0.047)**
* **Melhor qualidade média (5.97)**
* Açúcar residual moderado

**Interpretação:**
Esse cluster representa vinhos **mais equilibrados**, com boa acidez, menor densidade, álcool mais alto e **melhor qualidade geral**.

---

# **Cluster 3 – Vinhos intermediários**

* **Açúcar residual alto (6.86)**
* **SO₂ total intermediário (146)**
* **Álcool bom (10.34)**
* **Qualidade acima da média (5.89)**

**Interpretação:**
Grupo intermediário entre o Cluster 0 e 2: vinhos com certa doçura, mas ainda com boas características gerais.

---

# **2. Conclusão geral**

Com base no K-Means, encontramos **4 clusters distintos**, com perfis bem claros:

| Cluster | Perfil                                                       |
| ------- | ------------------------------------------------------------ |
| **0**   | Vinhos mais doces, densos, menores em qualidade              |
| **1**   | Vinhos secos, alcoólicos, porém com acidez volátil alta      |
| **2**   | Vinhos de melhor qualidade: álcool alto + equilíbrio químico |
| **3**   | Vinhos doces moderados com qualidade intermediária           |

O cluster **2** é o que melhor representa **vinhos de maior qualidade**.

O cluster **0** agrupa vinhos com pior avaliação química e sensorial.


### **Interpretação dos clusters**

Após aplicar o modelo K-Means ao dataset Wine Quality, foram identificados **quatro clusters principais**, com características químicas distintas. A análise das médias de cada grupo permite compreender perfis comuns entre os vinhos agrupados.

**Cluster 0**
Agrupa vinhos com **maior teor de açúcar residual** e **maior concentração de dióxido de enxofre total**, resultando em maior densidade. São vinhos mais doces, menos alcoólicos e com qualidade média inferior. Representa vinhos mais suaves e possivelmente de menor complexidade.

**Cluster 1**
Compreende vinhos **mais secos**, com menor açúcar residual, porém com **acidez volátil elevada** e maior presença de cloretos, o que pode impactar negativamente o aroma. Possui teor alcoólico elevado, mas qualidade apenas mediana. Representa vinhos mais agressivos aromaticamente.

**Cluster 2**
É o cluster com **melhor qualidade média** dentre os grupos. Apresenta **teor alcoólico mais alto**, baixa densidade e níveis moderados de açúcar residual e compostos sulfurosos. Indica vinhos mais equilibrados quimicamente, que tendem a apresentar melhor avaliação sensorial.

**Cluster 3**
Agrupa vinhos com **nível intermediário de açúcar residual** e teor alcoólico razoável, além de valores médios de compostos sulfurosos. Representa vinhos moderadamente doces, com qualidade acima da média, mas inferiores ao cluster 2.

**Conclusão:**
Os resultados mostram que o modelo de clusterização foi capaz de identificar segmentos claros dentro do conjunto de vinhos, distinguindo grupos por características químicas e de qualidade. O Cluster 2 se destaca por apresentar um perfil mais equilibrado e com maior qualidade, enquanto o Cluster 0 concentra vinhos mais doces e de qualidade inferior.
